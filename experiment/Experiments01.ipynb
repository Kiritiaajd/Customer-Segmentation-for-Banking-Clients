```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Step 1: Check for missing values in CustomerDOB
missing_dob = df['CustomerDOB'].isnull().sum()
print(f"Missing values in CustomerDOB: {missing_dob}")

# Step 2: Convert CustomerDOB to datetime format
df['CustomerDOB'] = pd.to_datetime(df['CustomerDOB'], errors='coerce')

# Step 3: Extract ages
current_year = pd.Timestamp.now().year
df['Age'] = current_year - df['CustomerDOB'].dt.year

# Step 4: Check descriptive statistics for age
print(df['Age'].describe())


# Step 6: Check for any invalid or extreme age values
invalid_ages = df[df['Age'] < 0]  # Negative ages
print(f"Invalid ages (negative values):\n{invalid_ages}")

extreme_ages = df[df['Age'] > 100]  # Extreme ages
print(f"Extreme ages (above 100):\n{extreme_ages}")
```

```python
# Summary statistics for CustAccountBalance
df['CustAccountBalance'].describe()
```

```python
# Calculate IQR
Q1 = df['CustAccountBalance'].quantile(0.25)
Q3 = df['CustAccountBalance'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['CustAccountBalance'] < lower_bound) | (df['CustAccountBalance'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")
```

```python
# Customers with negative balances
negative_balances = df[df['CustAccountBalance'] < 0]
print(f"Number of customers with negative balance: {len(negative_balances)}")

# Customers with zero balances
zero_balances = df[df['CustAccountBalance'] == 0]
print(f"Number of customers with zero balance: {len(zero_balances)}")
```

```python
# Calculate the 99th percentile
upper_limit = df['CustAccountBalance'].quantile(0.99)

# Cap the outliers
df['CustAccountBalance'] = np.where(df['CustAccountBalance'] > upper_limit, upper_limit, df['CustAccountBalance'])
```

```python
# Calculate IQR
Q1 = df['CustAccountBalance'].quantile(0.25)
Q3 = df['CustAccountBalance'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['CustAccountBalance'] >= lower_bound) & (df['CustAccountBalance'] <= upper_bound)]
```

```python
min_balance = df[df['CustAccountBalance'] > 0]['CustAccountBalance'].min()
df['CustAccountBalance'] = np.where(df['CustAccountBalance'] == 0, min_balance, df['CustAccountBalance'])
```
