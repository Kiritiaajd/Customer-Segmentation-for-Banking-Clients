## Odinal Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['CustGender'] = label_encoder.fit_transform(df['CustGender'])
df = pd.get_dummies(df, columns=['CustLocation'], drop_first=True)
df = pd.get_dummies(df, columns=['CustomerSegment'], drop_first=True)
```

```python
# Columns that need to be encoded (skip ones that are already encoded)
columns_to_encode = ['CustGender']  # You can add more columns if needed

# One-Hot Encoding only for the required columns that haven't been encoded yet
df = pd.get_dummies(df, columns=columns_to_encode, drop_first=True)

# Display the updated DataFrame
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_cols = ['TransactionAmount (INR)', 'FrequencyOfPurchasing', 'Age', 
                  'TransactionTime_in_hrs', 'MonthlyTransactionAverage', 
                  'RecencyDays', 'CustomerLifetimeValue', 'LoyaltyScore', 
                  'EstimatedYearlySpending']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_cols = ['TransactionAmount (INR)', 'FrequencyOfPurchasing', 'Age', 
                  'TransactionTime_in_hrs', 'MonthlyTransactionAverage']
for col in numerical_cols:
    sns.boxplot(df[col])
    plt.show()
```
