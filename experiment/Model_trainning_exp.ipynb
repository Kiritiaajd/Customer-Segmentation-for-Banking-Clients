```python
df_balanced = df.drop('HighSpender' , axis = True)
```

```python
df_balanced = df_balanced.drop('CustomerID' , axis= True)
df_balanced = df_balanced.drop('TransactionID' , axis= True)
df_balanced = df_balanced.drop('Year' , axis= True)
df_balanced = df_balanced.drop('CustLocation' , axis= True)
df_balanced = df_balanced.drop('TransactionTime_in_hrs' , axis= True)
```

```python
df = df_balanced
```

```python
# Savings Potential: CustAccountBalance - TotalSpending
df['SavingsPotential'] = df['CustAccountBalance'] - df['EstimatedYearlySpending']

# Interpret SavingsPotential
df['SavingsPotentialStatus'] = df['SavingsPotential'].apply(
    lambda x: 'Surplus' if x > 0 else 'Overspending'
)
```

```python
# Normalize each component for balanced scoring
df['RecencyScore'] = 1 / (1 + df['RecencyDays'])  # Normalize: Lower recency = Higher score
df['FrequencyScore'] = df['FrequencyOfPurchasing'] / df['FrequencyOfPurchasing'].max()  # Normalize frequency
df['MonetaryScore'] = df['EstimatedYearlySpending'] / df['EstimatedYearlySpending'].max()  # Normalize spending

# Weighted RFM Score: Customize weights as needed
df['RFM_Score'] = (df['RecencyScore'] * 0.4) + (df['FrequencyScore'] * 0.3) + (df['MonetaryScore'] * 0.3)
```

```python
df_copy = df[['CustAccountBalance', 'TransactionAmount (INR)', 'FrequencyOfPurchasing', 
              'CustomerLifetimeValue', 'MonthlyTransactionAverage', 'RecencyDays']]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select the relevant features from the DataFrame
df_copy = df[['CustAccountBalance', 'TransactionAmount (INR)', 'FrequencyOfPurchasing', 
              'CustomerLifetimeValue', 'MonthlyTransactionAverage', 'RecencyDays']]

# Normalize the data using StandardScaler (to scale features to have zero mean and unit variance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_copy)

# Apply KMeans clustering (let's assume we want 4 clusters for this example)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Lifestyle'] = kmeans.fit_predict(X_scaled)

# Map numerical cluster labels to customer lifestyle categories
lifestyle_map = {
    0: 'Low Spender',
    1: 'High Spender',
    2: 'Frequent Shopper',
    3: 'Occasional Shopper'
}

df['Lifestyle'] = df['Lifestyle'].map(lifestyle_map)

# Show the first few rows to verify the new column
print(df.head())
```
