```python
import pandas as pd

# Assuming your dataset is already loaded into a DataFrame df
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], format='%d/%m/%y')  # convert to datetime if not already

# Group by CustomerID and count the number of transactions
customer_freq = df.groupby('CustomerID')['TransactionDate'].count().reset_index()
customer_freq.rename(columns={'TransactionDate': 'FrequencyOfPurchasing'}, inplace=True)

# Merge this back with the original DataFrame
df = pd.merge(df, customer_freq, on='CustomerID', how='left')

# Now df contains the FrequencyOfPurchasing column
```

```python
df['CustomerDOB'] = pd.to_datetime(df['CustomerDOB'], errors='coerce')

# Calculate age (using reference date 2016-08-02)
reference_date = pd.Timestamp('2016-08-02')
df['Age'] = (reference_date - df['CustomerDOB']).dt.days // 365

# Display the DataFrame with the new Age column
```

```python
df['CustomerDOB'] = pd.to_datetime(df['CustomerDOB'], errors='coerce')
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], errors='coerce')

# Calculate age using TransactionDate as the reference
df['Age'] = (df['TransactionDate'] - df['CustomerDOB']).dt.days // 365

# Replace negative or zero ages with NaN to signify invalid DOB
df.loc[df['Age'] <= 0, 'Age'] = None

```
